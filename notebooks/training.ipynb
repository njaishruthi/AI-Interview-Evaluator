```python
import pandas as pd

df = pd.read_csv("../data/text/interview_questions.csv", encoding="latin1")

def evaluate_text(user_answer, expected_answer):
    score = 0

    if user_answer.lower() in expected_answer.lower():
        score += 4

    if len(user_answer.split()) > 5:
        score += 3

    keywords = ["because", "example", "use", "defined", "means"]
    if any(word in user_answer.lower() for word in keywords):
        score += 3

    return min(score, 10)

question = df.iloc[0]["Question"]
expected = df.iloc[0]["Answer"]

print("Question:", question)
print("\nExpected:", expected)

# 👇 manually give answer here
user_answer = str(input("enter your ans :"))

score = evaluate_text(user_answer, expected)

print("\nYour Answer:", user_answer)
print("Text Score:", score, "/10")
```
